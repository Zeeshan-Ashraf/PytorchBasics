# TORCH IMP POINTS
1. Tensor in torch have similar methods as pandas
2. torch only takes Tensor() obj
2. torch tensor operation results in tensor
3. 3 imp concept in Tensor - .shape(), dtype & device
4. float32 dtype is standard (integer params are less granular)
5. Tensor instead of List because Tensor has inbuilt gradient (differentiation) - requires_grad = True for doing d(Loss) / d(Weight) NOTE: we enable requires_grad for params (W & B) [why? coz: we need to find value of W & B by minimising d(Loss)/d(W) & d(Loss)/d(Bias) ]. zee!: if Loss = Y_pred-Y & Y_pred is = WX+B so to do d(Loss) / d(W) we need to do d(X)/d(W) then why not enable requires_grad for X
6. requires_grad = True tells pytorch for gradient to be calculated using chaining during backward pass & build graph (which tensor to have differentiation calculated) during forward pass
7. any new tensor created from a tensor with requires_grad = True will also have requires_grad = True
8. tensor has .grad_fn var which stores its source (parent) from where requires_grad = True request came

```python
'''
any operation with tensor results in tensor
tensor.randn also produce tensor
'''
import torch
import numpy as np
import torch.nn as nn

a = torch.tensor([1,2,3], dtype=torch.float32)
b = torch.randn(1,3, dtype=torch.float32, requires_grad=True)
b = a+b
type(a)
print(b,"||" , a, "||", b.item())
```
